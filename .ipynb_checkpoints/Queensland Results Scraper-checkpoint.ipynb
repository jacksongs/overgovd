```python
from bs4 import BeautifulSoup
import requests

soup1 = BeautifulSoup(requests.get('http://results.ecq.qld.gov.au/elections/local/LG2016/groupIndex.html').content,'lxml')
```

```python
links = soup1.find_all('a')
results = []
for link in links:
    if link.text.strip() == 'Councillor Results':
        results.append('http://results.ecq.qld.gov.au/elections/local/LG2016/'+link.get('href'))
```

```python
len(results)
```

```python
# Add extra Lockyer Valley election delayed due to death of mayor:
results.append('http://results.ecq.qld.gov.au/elections/local/LV2016/LockyerValleyRegionalCouncil/results/councillor/summary.html')
```

```python
collect = {}
for result in results:
    counter = 0

    soup2 = BeautifulSoup(requests.get(result).content,'lxml')
    tables = soup2.find_all('table')
    for table in tables:
        tab = table.find('td',class_='resultTableTitle')
        if tab != None:
            if tab.text.strip() == 'Declared Undivided Councils Details':
                trs = tab.parent.parent.find_all('tr',class_='contentRowOdd')
                for tr in trs:
                    tds = tr.find_all('td')
                    for td in tds:
                        if td.text.strip() != '' and td.text.strip() != '1':
                            if counter == 0:
                                council = td.text.strip()
                                collect[council] = []
                            else:
                                collect[council].append(td.text.strip())
                            counter += 1
                            
            else:
                if tab.text.strip() not in ['Results by Undivided Council List','Vote Type Summary']:
                    print(tab.text.strip())# Need to catch other electorates here
```

```python
# Now we have collected the results, let's get a count
reps = {}
for key in collect.keys():
    reps[key] = len(collect[key]) + 1 # Plus one for the mayor
```

```python
import json
with open('qldcounts.json', 'w') as fp:
    json.dump(reps, fp)
```
