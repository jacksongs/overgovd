```python
from fuzzywuzzy import process
import pandas as pd
abspop = pd.read_csv('ABS_ERP_LGA2015_08112016155122158.csv')
sapop = abspop[(abspop['LGA_2015']>39999)&(abspop['LGA_2015']<49000)]
sacounts = pd.read_csv('sacounts.csv')
```

```python
# The goal will be to receive two dataframe of equal length, name the columns subject to the match
# It produces the best joined table with a score for each


def bestfuzzy(d1,d2,c1,c2,depth=5):
    # first off, get a list of depth-length matches for each 
    matches = []
    for key in d1[c1].tolist():
        for result in process.extract(key,d2[c2].tolist(),limit=depth):
            matches.append([key,result[0],result[1]])
    test = pd.DataFrame(matches)
    common = pd.DataFrame(test.groupby(1).size().sort_values())
    return pd.merge(test,common,left_on=1,right_index=True).sort_values([2,'0_y'])
    
test = bestfuzzy(sapop,sacounts,'Region','Council',1)
```
