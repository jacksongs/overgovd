```python
nswdata = []
for l in links:
    page = requests.get('https://www.olg.nsw.gov.au'+l[0])
    souper = BeautifulSoup(page.content,'lxml')
    content = souper.find_all('div',class_='content')
    ps = content[1].find_all('p')
    row = [l[1],l[2]]
    for p in ps:
        
        try:
            if p.findChildren()[0].get('class')[0] == 'lwdir_detail_label_b':
                
                if p.findChildren()[0].text.strip() == 'COUNCILLORS:':
                    row.append(p.findChildren()[1].text.replace("(","").replace(")",""))
                if p.findChildren()[0].text.strip() == 'Area (sq Km):':
                    row.append(p.findChildren()[1].text.replace("(","").replace(")",""))
                if p.findChildren()[0].text.strip() == 'Population:':
                    row.append(p.findChildren()[1].text.replace("(","").replace(")",""))
            
        except:
            pass
    nswdata.append(row)
```
